```python
from calculations.cian import get_cian_results
from calculations.wildberries import get_wildberries_results
from calculations.habr import get_habr_results
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cian_results = get_cian_results(number_launches=30)
x = []
y1 = [] # lxml.html
y2 = [] # bs4 
y3 = [] # scrapy
for res in cian_results:
    x.append(res['test_id'].split('_')[1])
    y1.append(res['lxml_time'])
    y2.append(res['bs4_time'])
    y3.append(res['scrapy_time'])

# Prepare Data

mycolors = ['tab:red', 'tab:blue', 'tab:green']      
columns = ['lxml', 'bs4', 'scrapy']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(16,9), dpi= 200)
ax.fill_between(x, y1=y1, y2=0, label=columns[0], alpha=0.4, color=mycolors[0], linewidth=2)
ax.fill_between(x, y1=y2, y2=0, label=columns[1], alpha=0.4, color=mycolors[1], linewidth=2)
ax.fill_between(x, y1=y3, y2=0, label=columns[2], alpha=0.4, color=mycolors[2], linewidth=2)

# Decorations
ax.set_title('Cian', fontsize=18)
ax.set(ylim=[0, 0.000001])
ax.legend(loc='best', fontsize=12)
plt.xticks(x, fontsize=10, horizontalalignment='center')
plt.yticks(np.arange(0, 0.000003,0.0000001), fontsize=10)
plt.xlim(-1, x[-1])

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.show()
```
