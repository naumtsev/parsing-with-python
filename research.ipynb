```python
from calculations.cian import get_cian_results
from calculations.wildberries import get_wildberries_results
from calculations.habr import get_habr_results
```

```python
import matplotlib.pyplot as plt
import numpy as np
cian_results = get_cian_results(group_size=3, number_launches=100)
wildberries_results = get_wildberries_results(group_size=10, number_launches=6)
habr_results = get_habr_results(group_size=15, number_launches=1)
```

```python
x = []
y1 = [] # lxml.html
y2 = [] # bs4 
y3 = [] # scrapy
for res in cian_results:
    x.append(res['test_id'].split('_')[1])
    y1.append(res['lxml_time'])
    y2.append(res['bs4_time'])
    y3.append(res['scrapy_time'])

# Prepare Data

mycolors = ['tab:red', 'tab:blue', 'tab:green']      
columns = ['lxml', 'bs4', 'scrapy']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(16,9), dpi= 200)
ax.fill_between(x, y1=y1, y2=0, label=columns[0], alpha=0.4, color=mycolors[0], linewidth=2)
ax.fill_between(x, y1=y2, y2=0, label=columns[1], alpha=0.4, color=mycolors[1], linewidth=2)
ax.fill_between(x, y1=y3, y2=0, label=columns[2], alpha=0.4, color=mycolors[2], linewidth=2)

# Decorations
ax.set_title('Cian', fontsize=18)
ax.set(ylim=[0, 0.0000001])
ax.legend(loc='best', fontsize=12)
plt.xticks(x, fontsize=10, horizontalalignment='center')
plt.yticks(np.arange(0, 0.0000015,0.0000001), fontsize=10)
plt.xlim(-1, x[-1])

for y in np.arange(0, 0.0000012, 0.0000001):    
    plt.hlines(y, xmin=0, xmax=len(x), colors='black', alpha=0.3, linestyles="--", lw=0.5)

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.show()
```

```python
results = [(cian_results, 'Cian'), (wildberries_results, 'Wildberries'), (habr_results, 'Habr')]
mycolors = ['#FB717E', '#3D9AD1', '#A6F16C']      
gridsize = (1, len(results))
fig = plt.figure(figsize=(16, 5))

for i in range(len(results)):
    points = {'lxml': 0, 'bs4': 0, 'scrapy': 0}
    for res in results[i][0]:
        times = [(res['lxml_time'], 'lxml'), (res['bs4_time'], 'bs4'), (res['scrapy_time'], 'scrapy')]
        times.sort()
#         points[times[0][1]] += 1
        for t, p in times:
            if t / times[0][0] <= 1.03:
                points[p] += 1

    labels = []
    sizes = []
    for parser_name in points:
        labels.append(parser_name)
        sizes.append(points[parser_name])
        
    ax = plt.subplot2grid(gridsize, (0, i))
    ax.set_title(results[i][1], fontsize=25)
    ax.pie(sizes, labels=labels, colors=mycolors, autopct='%1.2f%%')
    ax.axis('equal')
    ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0), fontsize=12)

plt.show()
```
